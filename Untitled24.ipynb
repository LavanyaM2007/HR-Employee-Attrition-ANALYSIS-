```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

df.head()
```

```python
print("Shape:", df.shape)

df.info()

df.describe()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
df.shape
attrition_rate = df['Attrition'].value_counts()

print(attrition_rate)
```

```python
plt.figure(figsize=(6,6))

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Employee Attrition Rate")
plt.ylabel('')
plt.show()
```

```python
dept_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

print(dept_attrition)
```

```python
dept_attrition.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Employees")
plt.show()
```

```python
salary_attrition = df.groupby('Attrition')['MonthlyIncome'].mean()

print(salary_attrition)
plt.figure(figsize=(6,4))

sns.barplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.title("Average Salary vs Attrition")
plt.show()
```

```python
exp_attrition = df.groupby('Attrition')['TotalWorkingYears'].mean()

print(exp_attrition)
plt.figure(figsize=(6,4))

sns.barplot(
    x='Attrition',
    y='TotalWorkingYears',
    data=df
)

plt.title("Experience vs Attrition")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='JobSatisfaction',
    hue='Attrition',
    data=df
)

plt.title("Job Satisfaction vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.title("OverTime vs Attrition")
plt.show()
```

```python
print("PROJECT INSIGHTS")

print("""
1. Employees doing overtime show higher attrition.
2. Low job satisfaction employees leave more.
3. Employees with lower salary tend to leave.
4. Some departments have higher attrition rates.
5. Less experienced employees are more likely to leave.
""")
```
